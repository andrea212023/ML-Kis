1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

# Define the function
x = symbols('x')
f_x = 10 * x - 5

# Calculate the derivative of the function
f_x_derivative = diff(f_x, x)
print(f"Derivative of f(x): {f_x_derivative}")

# Convert functions to numerical functions
f = lambda x: 10 * x - 5
# The derivative is constant, but we need to return an array
# with the same shape as the input for plotting
f_prime = lambda x: np.full(x.shape, 10)  # Corrected line

# Generate values for plotting
x_values = np.linspace(0, 10, 100)
f_values = f(x_values)
f_prime_values = f_prime(x_values)

# Plot the function and its derivative
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label="f(x) = 10x - 5", color='blue')
plt.plot(x_values, f_prime_values, label="f'(x) = 10", color='red', linestyle='--')
plt.xlabel("Hours of Study")
plt.ylabel("Grade")
plt.title("Dependency of Grades on Study Hours and its Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

**Analysis:**

The function f(x) = 10x - 5 describes a linear increase in grade with each hour spent studying.

The derivative f'(x) = 10 indicates a constant rate of change: each additional hour of study increases the grade by 10 points.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

# Define the symbols
x = sp.symbols('x')

# Define the functions
f1 = sp.sin(x)
f2 = sp.ln(x)

# Calculate the derivatives
f1_derivative = sp.diff(f1, x)
f2_derivative = sp.diff(f2, x)

# Print the derivatives
print(f"The derivative of sin(x) is: {f1_derivative}")
print(f"The derivative of ln(x) is: {f2_derivative}")

import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x_vals = np.linspace(0.1, 10, 400)

# Define the functions
f1_vals = np.sin(x_vals)
f2_vals = np.log(x_vals)

# Define the derivatives
f1_derivative_vals = np.cos(x_vals)
f2_derivative_vals = 1 / x_vals

# Create the plots
plt.figure(figsize=(10, 6))

# Plot f1 and its derivative
plt.subplot(2, 2, 1)
plt.plot(x_vals, f1_vals, label='sin(x)')
plt.title('sin(x)')
plt.subplot(2, 2, 2)
plt.plot(x_vals, f1_derivative_vals, label="Derivative of sin(x)", color='r')
plt.title("Derivative of sin(x)")

# Plot f2 and its derivative
plt.subplot(2, 2, 3)
plt.plot(x_vals, f2_vals, label='ln(x)')
plt.title('ln(x)')
plt.subplot(2, 2, 4)
plt.plot(x_vals, f2_derivative_vals, label="Derivative of ln(x)", color='r')
plt.title("Derivative of ln(x)")

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the variable and function
x = sp.symbols('x')
f = x**2 - 5*x + 1

# Calculate the derivative
f_derivative = sp.diff(f, x)

# Lambdify the functions for numerical evaluation
f_lambdified = sp.lambdify(x, f, 'numpy')
f_derivative_lambdified = sp.lambdify(x, f_derivative, 'numpy')

# Define the range for x
x_vals = np.linspace(0, 100, 400)

# Evaluate the functions
f_vals = f_lambdified(x_vals)
f_derivative_vals = f_derivative_lambdified(x_vals)

# Create the plots
plt.figure(figsize=(10, 6))

# Plot f(x)
plt.subplot(2, 1, 1)
plt.plot(x_vals, f_vals, label='f(x) = x^2 - 5x + 1')
plt.title('f(x) = x^2 - 5x + 1')
plt.legend()

# Plot f'(x)
plt.subplot(2, 1, 2)
plt.plot(x_vals, f_derivative_vals, label="f'(x) = 2x - 5", color='r')
plt.title("Derivative of f(x) = x^2 - 5x + 1")
plt.legend()

plt.tight_layout()
plt.show()
```
